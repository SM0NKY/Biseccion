```python
from typing import Optional, Any
import time
import pandas as pd
from typing import Literal
import pandas as pd
```

```python
def funcion_cubica(x:int|float) -> Optional[int|Any]:
    """
    Permite evaluar una funcion cúbica fija

    Parameters
    ----------
    x : int | float
        Es el valor que se va a evaluar  

    """
    
    return x**3 - 4*x**2 + x + 6
```

```python
class Raices(object):
    """ 

    Parameters
    ----------
    eta  : float | int
        Permite conocer el valor del error que se puede tener

    Atributes
    ----------
    self.xl : int | float
        Este valor le pertenece al limite mas alto 
    self.xu : int | float
        Este valor le pertenece al limite mas bajo
    self.Et : float | int
        Es el valor inicial del porcentaje de error
    self.iteration : int
        Numero de iteraciones realizadas dentro del ciclo
        
    """
    
    def __init__(self, eta:float|int = 0.01):
        self.eta:float = eta # %
        self.xl:int|float = 2.99
        self.xu:int|float = 1
        self.true_value:float = 2
        self.xr:int = 0
        self.Et:float|int = 100 # %
        self.iteration:int = 0
        

    def xr_value(self) -> Optional[int|float]:
        """
        Utiliza el metodo de biseccion para aproximarse a la raíz
        
        Parameters
        ----------
        `None`
        """
        self.xr = (self.xl + self.xu)/2 
        return self.xr # Se utiliza el valor de la regla 
    
    def new_value(self) -> Optional[int|float]:
        """
        Utiliza el metodo de la regla falsa para aproximarse a la raiz

        Parameters
        ----------
        `None`
        """
        self.xr = self.xu - (funcion_cubica(self.xu)*(self.xl-self.xu))/(funcion_cubica(self.xl)- funcion_cubica(self.xu))
        return self.xr #Utiliza el metodo de la regla falsa para obtener el valor de xr

    def check_rule(self, method:Literal[0,1]) -> Optional[dict]:
        """
        Se Evaluan los valores para calcular las raices
        
        Parameters
        ----------
        method : Literal[0,1]
            Son los valores utilizados para indicar el metodo a utilizar
            Para bisección (0)
            Para regla falsa (1)
        """
        
        self.iter:list = []
        self.error:list = []
        self.value:list = []
        #Aqui se agrega el ciclo while para esperar a que el error sea minimo #
        while self.Et > self.eta:
            
            self.iteration += 1
            if method == 0:
                self.xr = self.xr_value()
            elif method == 1:
                self.xr = self.new_value()
            
            producto = funcion_cubica(self.xr)*funcion_cubica(self.xl) # Se obtiene el producto para ver que signo tiene

            
            self.Et = (abs(self.true_value - self.xr)/self.xr) * 100 # Calculo del error relativo porcentual
            
            
            self.iter.append(self.iteration)
            self.value.append(self.xr)
            self.error.append(self.Et)

            if producto > 0: # Verificar si se encuentra el valor del producto es negativo o no, para escoger un valor y poder acercarse a la raíz
                self.xl = self.xr
            elif producto < 0:
                self.xu = self.xr
            else:
                break
                
        self.values_tabelle:dict = {
            'Iteration': self.iter,
            'Error': self.error,
            'Value': self.value,
        }

        return self.values_tabelle
    
```

```python
raices:object|Raices = Raices()
```

```python
data = raices.check_rule(method = 0)
```
